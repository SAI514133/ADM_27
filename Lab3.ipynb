```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([0,6])
ypoints = np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints, 'o')
plt.show()
```

```python
ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints, 'o:r')
plt.show()
```

```python
plt.plot(ypoints,marker='o',ms=20,mec='royalblue',mfc='yellow')
plt.plot(ypoints,ls=':')
plt.plot(ypoints,color='r')
plt.plot(ypoints,linewidth='20.5')
plt.show()
```

```python
x1=np.array([0,1,2,3])
y1=np.array([3,8,1,10])
x2=np.array([0,1,2,3])
y2=np.array([6,2,7,11])
plt.plot(x1,y1,x2,y2)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

```python
plt.grid()
plt.grid(axis='x')
plt.grid(axis='y')
```
